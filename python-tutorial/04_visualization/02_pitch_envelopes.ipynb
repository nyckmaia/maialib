## Why This Matters

While piano rolls show every individual note, pitch envelopes show the **overall trajectory** of pitch over time. Think of it as:

- **Piano Roll** = Detailed map showing every street and building
- **Pitch Envelope** = Elevation profile showing the terrain's overall shape

Pitch envelopes are especially useful for:

- **Understanding melodic shape**: Does the melody rise, fall, or stay level?
- **Comparing tessitura**: Which instruments play in high vs. low registers?
- **Tracking development**: How does pitch range change throughout a piece?
- **Identifying climax points**: Where are the highest/lowest pitches?
- **Analyzing voice leading**: How do different parts move in relation to each other?

This visualization is particularly valuable for:
- Music analysts studying form and structure
- Composers learning from masterworks
- Teachers explaining melodic concepts
- Researchers studying pitch organization

## Step 2: Your First Pitch Envelope

Let's start with a simple melody and see its pitch envelope.

````python
# Load Bach's Cello Suite No. 1 (single melodic line)
bach = ml.Score('Bach_Cello_Suite_1.mxl')

print("Score Information:")
print(f"  Title: {bach.getWorkTitle()}")
print(f"  Composer: {bach.getComposer()}")
print(f"  Measures: {bach.getNumMeasures()}")

# Create a pitch envelope visualization
ml.plotScorePitchEnvelope(bach)

print("\n📈 Pitch envelope created!")
````

**What You're Seeing:**

- **X-axis**: Time in measures
- **Y-axis**: Pitch (MIDI number - higher = higher pitch)
- **Line plot**: Shows the pitch of each note as it occurs
- **Overall shape**: The melodic contour of the entire piece

**Musical Insight**: Notice:
- The **regular wave pattern** - this reflects the arpeggiated nature of the Prelude
- The **gradual rise and fall** - Bach creates long-term motion even within constant 16th notes
- The **tessitura** (average range) - most notes cluster around D4-A4 (the cello's comfortable middle range)
- **Peak moments** - the highest pitches create climax points

## Step 3: Comparing Piano Roll vs. Pitch Envelope

Let's visualize the same piece both ways to understand the difference.

````python
# Load a score
score = ml.Score('Bach_Cello_Suite_1.mxl')

# Get just the first 8 measures for clarity
part = score.getPart(0)
excerpt = ml.Score()
excerpt_part = ml.Part()
excerpt_part.setPartName("Cello (mm. 1-8)")

for i in range(8):
    excerpt_part.addMeasure(part.getMeasure(i))

excerpt.addPart(excerpt_part)

# Piano roll
print("Piano Roll (shows every note as a rectangle):")
ml.plotPianoRoll(excerpt)

# Pitch envelope
print("\nPitch Envelope (shows melodic contour):")
ml.plotScorePitchEnvelope(excerpt)

print("\n📊 Notice the difference:")
print("  - Piano roll: Shows duration and exact timing")
print("  - Pitch envelope: Shows overall melodic shape and trajectory")
````

**Key Differences:**

| Aspect | Piano Roll | Pitch Envelope |
|--------|-----------|----------------|
| **Focus** | Individual notes | Overall contour |
| **Duration** | Visible (rectangle width) | Not shown |
| **Pattern recognition** | Detailed, specific | General, structural |
| **Best for** | Rhythm, texture, density | Melodic shape, range, tessitura |
| **Information density** | High (can be overwhelming) | Low (clearer overview) |

Use **piano rolls** when you need detail. Use **pitch envelopes** when you need overview.

## Step 4: Multi-Part Pitch Envelopes

Now let's see how different instruments move in relation to each other.

````python
# Load Mozart Requiem (multiple vocal and instrumental parts)
mozart = ml.Score('Mozart_Requiem_Introitus.mxl')

print("Score Information:")
print(f"  Title: {mozart.getWorkTitle()}")
print(f"  Parts: {mozart.getNumParts()}")

# List parts
print("\nParts in the score:")
for i in range(min(8, mozart.getNumParts())):  # Show first 8 parts
    part = mozart.getPart(i)
    print(f"  {i}: {part.getPartName()}")

# Create pitch envelope for all parts
ml.plotScorePitchEnvelope(mozart)

print("\n📈 Multi-part pitch envelope created!")
````

**What You're Seeing:**

- **Multiple colored lines**: Each part/instrument has its own color
- **Vertical stratification**: Different parts occupy different pitch ranges
- **Voice leading**: How parts move in relation to each other
- **Convergence/divergence**: When parts come together or spread apart

**Musical Analysis:**

- **Sopranos** (top line): Highest pitches, often the melody
- **Altos** (upper-middle): Supporting harmony
- **Tenors** (lower-middle): Inner voices
- **Basses** (bottom): Foundation and harmonic root
- **Instruments** (various colors): Double or support the vocal lines

Notice how Mozart maintains clear separation between voice types while creating moments of overlap for special effects!

## Step 5: Analyzing Tessitura and Range

Let's extract numerical data about pitch ranges from the visualization.

````python
# Load Beethoven Symphony No. 5
beethoven = ml.Score('Beethoven_Symphony_5_mov_1.xml')

# Analyze each part's pitch characteristics
print("Pitch Analysis by Part:")
print("\n{:<20} {:<12} {:<12} {:<12} {:<12}".format(
    "Part", "Lowest", "Highest", "Range", "Average"))
print("="*70)

for i in range(min(5, beethoven.getNumParts())):  # First 5 parts
    part = beethoven.getPart(i)
    
    # Get all notes from this part
    all_notes = []
    for m in range(part.getNumMeasures()):
        measure = part.getMeasure(m)
        notes = measure.getNotes()
        all_notes.extend(notes)
    
    if len(all_notes) > 0:
        # Calculate statistics
        midi_values = [note.getMIDI() for note in all_notes]
        lowest = min(midi_values)
        highest = max(midi_values)
        range_semitones = highest - lowest
        average = sum(midi_values) / len(midi_values)
        
        # Convert to pitch names
        lowest_pitch = ml.midiToPitch(lowest)
        highest_pitch = ml.midiToPitch(highest)
        average_pitch = ml.midiToPitch(int(average))
        
        print("{:<20} {:<12} {:<12} {:<12} {:<12}".format(
            part.getPartName()[:20],
            f"{lowest_pitch} ({lowest})",
            f"{highest_pitch} ({highest})",
            f"{range_semitones} st",
            f"{average_pitch} ({average:.1f})"
        ))

# Visualize the pitch envelopes
print("\nPitch Envelope Visualization:")
ml.plotScorePitchEnvelope(beethoven)
````

**Understanding the Data:**

- **Range**: The span from lowest to highest note (in semitones)
  - Small range (< 12): Limited, focused writing
  - Medium range (12-24): Comfortable, typical
  - Large range (> 24): Virtuosic, challenging

- **Average (Tessitura)**: The typical pitch center
  - High tessitura: Bright, intense, sometimes tiring
  - Middle tessitura: Comfortable, sustainable
  - Low tessitura: Dark, powerful, sometimes muddy

**Compositional Insight**: Beethoven typically writes:
- Violins: High range, wide compass
- Violas: Middle range, moderate compass
- Cellos/Basses: Low range, sometimes very wide compass

This creates the characteristic orchestral blend!

## Step 6: Tracking Melodic Contour

Let's analyze the shape of a melody by looking at its pitch envelope.

````python
# Load Bach Prelude
bach_prelude = ml.Score('Bach_Prelude_1_BWV_846.mxl')

# Get the piano part
piano = bach_prelude.getPart(0)

# Analyze overall contour
all_notes = []
for m in range(piano.getNumMeasures()):
    measure = piano.getMeasure(m)
    notes = measure.getNotes()
    all_notes.extend(notes)

# Get MIDI values
midi_values = [note.getMIDI() for note in all_notes]

# Find peaks (highest points)
max_midi = max(midi_values)
max_index = midi_values.index(max_midi)

# Find valleys (lowest points)
min_midi = min(midi_values)
min_index = midi_values.index(min_midi)

print("Melodic Contour Analysis:")
print(f"  Total notes: {len(all_notes)}")
print(f"  Overall range: {min_midi} to {max_midi} ({max_midi - min_midi} semitones)")
print(f"  Highest point: {ml.midiToPitch(max_midi)} at note {max_index}")
print(f"  Lowest point: {ml.midiToPitch(min_midi)} at note {min_index}")
print(f"  Climax position: {(max_index / len(all_notes)) * 100:.1f}% through the piece")

# Visualize
ml.plotScorePitchEnvelope(bach_prelude)

print("\n🎵 Observe:")
print("  - Where is the highest pitch? (Climax)")
print("  - Does the melody generally rise, fall, or arch?")
print("  - Are there repeated patterns in the contour?")
````

**Contour Analysis Techniques:**

Common melodic shapes you might see:

1. **Arch**: Rises to a peak, then descends (very common, natural)
2. **Ascending**: Generally rises throughout (creates excitement)
3. **Descending**: Generally falls (creates resolution/relaxation)
4. **Terraced**: Steps up or down in plateaus
5. **Wave**: Repeated rise-and-fall patterns (like Bach arpeggios)

The **climax position** is important:
- Early (< 33%): Builds energy, then releases it
- Middle (33-66%): Classic balanced form
- Late (> 66%): Builds to final climax (dramatic)

## Step 7: Comparing Pitch Evolution Across Pieces

Let's compare how different composers use pitch over time.

````python
# Load two contrasting pieces
chopin = ml.Score('Chopin_Fantasie_Impromptu.mxl')
strauss = ml.Score('Strauss_Blue_Danube_Waltz.mxl')

print("Chopin Fantaisie-Impromptu:")
chopin_df = chopin.toDataFrame()
print(f"  Range: {chopin_df['midi'].min()} - {chopin_df['midi'].max()} ({chopin_df['midi'].max() - chopin_df['midi'].min()} semitones)")
print(f"  Average pitch: {chopin_df['midi'].mean():.1f} ({ml.midiToPitch(int(chopin_df['midi'].mean()))})")
print(f"  Std deviation: {chopin_df['midi'].std():.1f} semitones")

ml.plotScorePitchEnvelope(chopin)

print("\nStrauss Blue Danube Waltz:")
strauss_df = strauss.toDataFrame()
print(f"  Range: {strauss_df['midi'].min()} - {strauss_df['midi'].max()} ({strauss_df['midi'].max() - strauss_df['midi'].min()} semitones)")
print(f"  Average pitch: {strauss_df['midi'].mean():.1f} ({ml.midiToPitch(int(strauss_df['midi'].mean()))})")
print(f"  Std deviation: {strauss_df['midi'].std():.1f} semitones")

ml.plotScorePitchEnvelope(strauss)
````

**Comparative Analysis:**

**Chopin** (Romantic piano music):
- Very wide overall range (uses full keyboard)
- High standard deviation (lots of pitch variety)
- Dramatic leaps between registers
- Complex, irregular contours

**Strauss** (Classical orchestral waltz):
- Wide range (orchestra spans many octaves)
- Clear stratification (melody vs. accompaniment)
- More regular, predictable contours
- Dance-like patterns visible in the envelope

**Standard Deviation Tells You:**
- Low (< 12): Narrow range, consistent tessitura
- Medium (12-24): Typical variety
- High (> 24): Wide-ranging, dramatic pitch usage

## Step 8: Advanced - Creating Custom Pitch Statistics

Let's calculate additional metrics to describe pitch behavior.

````python
import pandas as pd

def analyze_pitch_envelope(score):
    """Calculate detailed pitch envelope statistics."""
    
    df = score.toDataFrame()
    
    # Basic statistics
    stats = {
        'total_notes': len(df),
        'min_pitch': df['midi'].min(),
        'max_pitch': df['midi'].max(),
        'range': df['midi'].max() - df['midi'].min(),
        'mean_pitch': df['midi'].mean(),
        'median_pitch': df['midi'].median(),
        'std_pitch': df['midi'].std(),
    }
    
    # Calculate pitch motion
    pitch_changes = df['midi'].diff().dropna()
    stats['avg_interval'] = abs(pitch_changes).mean()
    stats['max_leap'] = abs(pitch_changes).max()
    
    # Direction analysis
    stats['ascending_intervals'] = (pitch_changes > 0).sum()
    stats['descending_intervals'] = (pitch_changes < 0).sum()
    stats['repeated_notes'] = (pitch_changes == 0).sum()
    
    return stats

# Analyze a score
score = ml.Score('Bach_Cello_Suite_1.mxl')
stats = analyze_pitch_envelope(score)

print("Pitch Envelope Statistics:")
print(f"\nRange and Tessitura:")
print(f"  Min pitch: {ml.midiToPitch(int(stats['min_pitch']))} ({stats['min_pitch']})")
print(f"  Max pitch: {ml.midiToPitch(int(stats['max_pitch']))} ({stats['max_pitch']})")
print(f"  Range: {stats['range']} semitones")
print(f"  Mean pitch: {ml.midiToPitch(int(stats['mean_pitch']))} ({stats['mean_pitch']:.1f})")
print(f"  Pitch std dev: {stats['std_pitch']:.1f} semitones")

print(f"\nMelodic Motion:")
print(f"  Average interval size: {stats['avg_interval']:.2f} semitones")
print(f"  Largest leap: {stats['max_leap']:.0f} semitones")
print(f"  Ascending intervals: {stats['ascending_intervals']} ({stats['ascending_intervals']/stats['total_notes']*100:.1f}%)")
print(f"  Descending intervals: {stats['descending_intervals']} ({stats['descending_intervals']/stats['total_notes']*100:.1f}%)")
print(f"  Repeated notes: {stats['repeated_notes']} ({stats['repeated_notes']/stats['total_notes']*100:.1f}%)")

# Visualize
ml.plotScorePitchEnvelope(score)
````

**Interpreting Advanced Statistics:**

**Average Interval Size:**
- Small (< 2): Stepwise, conjunct motion (singable, smooth)
- Medium (2-4): Mix of steps and small leaps (typical)
- Large (> 4): Disjunct, angular motion (dramatic, challenging)

**Directional Balance:**
- Equal ascending/descending: Balanced, arch-like contours
- More ascending: Overall rising trajectory
- More descending: Overall falling trajectory

**Repeated Notes:**
- High %: Focused on rhythm, static harmony
- Low %: Continuous melodic flow

def find_climax(score):
    """Find the climax point (highest pitch) in a score."""
    df = score.toDataFrame()
    
    # Find maximum pitch
    max_idx = df['midi'].idxmax()
    climax_row = df.loc[max_idx]
    
    # Calculate position
    position_pct = (max_idx / len(df)) * 100
    
    print(f"Climax Analysis:")
    print(f"  Highest pitch: {climax_row['pitch']} (MIDI {climax_row['midi']})")
    print(f"  Measure: {climax_row['measure']}")
    print(f"  Position: {position_pct:.1f}% through the piece")
    
    if position_pct < 33:
        print(f"  Placement: Early climax (builds energy early)")
    elif position_pct < 66:
        print(f"  Placement: Middle climax (balanced form)")
    else:
        print(f"  Placement: Late climax (builds to finale)")
    
    return climax_row
