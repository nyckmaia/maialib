## Why This Matters

Piano roll visualizations are one of the most intuitive ways to see music. They show:

- **When notes happen** (horizontal position = time)
- **Which notes play** (vertical position = pitch)
- **How long notes last** (rectangle width = duration)

This format is familiar to anyone who has used a MIDI editor, digital audio workstation (DAW), or even seen a player piano roll. It's an excellent tool for:

- Understanding melodic contours and ranges
- Identifying patterns and repetitions
- Analyzing texture and density
- Comparing different sections or pieces
- Creating publication-quality figures

## Step 2: Your First Piano Roll

Let's create the simplest possible piano roll - a single line melody.

````python
# Load Bach's Cello Suite No. 1 (single melodic line)
bach = ml.Score('Bach_Cello_Suite_1.mxl')

print("Score Information:")
print(f"  Title: {bach.getWorkTitle()}")
print(f"  Composer: {bach.getComposer()}")
print(f"  Measures: {bach.getNumMeasures()}")
print(f"  Notes: {bach.getNumNotes()}")

# Create a piano roll visualization
ml.plotPianoRoll(bach)

print("\n🎨 Piano roll created!")
````

**What You're Seeing:**

- **X-axis**: Time in measures (each vertical line is a bar line)
- **Y-axis**: Pitch (MIDI number - higher numbers = higher notes)
- **Blue rectangles**: Individual notes (length = duration)
- **Overall pattern**: You can see the famous opening arpeggio pattern repeating

**Musical Insight**: Notice the regular, flowing pattern - this is what gives the Prelude its characteristic rippling effect. The notes cluster around certain pitches (D, G, B) which are the open strings of the cello and the notes of the G major chord.

## Step 3: Multi-Part Piano Roll

Now let's visualize a more complex score with multiple parts (instruments).

````python
# Load Mozart Requiem (choir + orchestra)
mozart = ml.Score('Mozart_Requiem_Introitus.mxl')

print("Score Information:")
print(f"  Title: {mozart.getWorkTitle()}")
print(f"  Composer: {mozart.getComposer()}")
print(f"  Parts: {mozart.getNumParts()}")
print(f"  Measures: {mozart.getNumMeasures()}")

# List all parts
print("\nParts in the score:")
for i in range(mozart.getNumParts()):
    part = mozart.getPart(i)
    print(f"  {i}: {part.getPartName()}")

# Create piano roll of the entire score
ml.plotPianoRoll(mozart)

print("\n🎨 Multi-part piano roll created!")
````

**What You're Seeing:**

- **Different colors**: Each part (instrument/voice) has a different color
- **Vertical spread**: Notice how different instruments occupy different pitch ranges
- **Texture**: Dense sections = more instruments playing together
- **Gaps**: Silences or rests in the music

**Musical Insight**: You can see the orchestral texture clearly:
- **High notes**: Violins, sopranos
- **Middle range**: Violas, altos, tenors
- **Low notes**: Cellos, basses, bass voice

Notice how Mozart creates dramatic moments by having all parts play together (dense vertical sections) versus solo passages (thin vertical lines).

## Step 4: Visualizing a Single Part

Sometimes you want to focus on just one instrument or voice. Let's extract and visualize a single part.

````python
# Load Beethoven Symphony No. 5
beethoven = ml.Score('Beethoven_Symphony_5_mov_1.xml')

print("Score Information:")
print(f"  Title: {beethoven.getWorkTitle()}")
print(f"  Parts: {beethoven.getNumParts()}")

# Get the first violin part
violin1 = beethoven.getPart(0)
print(f"\nVisualizing: {violin1.getPartName()}")

# Create a score from just this part
violin1_score = ml.Score()
violin1_score.addPart(violin1)

# Visualize just the first violin
ml.plotPianoRoll(violin1_score)

print("\n🎨 Single-part piano roll created!")
````

**Musical Insight**: By isolating the first violin, you can clearly see:

- The famous "fate knocking at the door" motif (short-short-short-long)
- Repeated patterns and sequences
- The melodic contour (shape of the melody)
- Moments of rest versus continuous playing

This is incredibly useful for studying how a composer develops themes or for understanding the role of each instrument in the texture.

## Step 5: Visualizing a Specific Section

What if you only want to see measures 10-20? Let's learn how to extract and visualize specific sections.

````python
# Load Bach Prelude No. 1
bach_prelude = ml.Score('Bach_Prelude_1_BWV_846.mxl')

print(f"Total measures: {bach_prelude.getNumMeasures()}")

# Create a new score with just measures 1-8 (the opening)
opening = ml.Score()

# Get the piano part
piano_part = bach_prelude.getPart(0)

# Create a new part for the excerpt
excerpt_part = ml.Part()
excerpt_part.setPartName("Piano (measures 1-8)")

# Copy measures 0-7 (0-indexed!)
for i in range(8):
    measure = piano_part.getMeasure(i)
    excerpt_part.addMeasure(measure)

opening.addPart(excerpt_part)

# Visualize the opening
ml.plotPianoRoll(opening)

print("\n🎨 Section piano roll created (measures 1-8)!")
````

**Musical Insight**: This famous opening shows Bach's brilliant use of broken chords (arpeggios). Each measure contains a different harmony:

- Notice the regular, repeating rhythmic pattern
- See how the pitch pattern changes slightly each measure
- The visual symmetry reflects the musical symmetry

This visualization makes it easy to see how Bach creates variation within repetition - a key compositional technique.

## Step 6: Comparing Two Pieces Side by Side

Let's create piano rolls for two different pieces to compare their characteristics.

````python
# Load two contrasting works
chopin = ml.Score('Chopin_Fantasie_Impromptu.mxl')
strauss = ml.Score('Strauss_Blue_Danube_Waltz.mxl')

print("Chopin Fantaisie-Impromptu:")
print(f"  Measures: {chopin.getNumMeasures()}")
print(f"  Notes: {chopin.getNumNotes()}")

print("\nStrauss Blue Danube Waltz:")
print(f"  Measures: {strauss.getNumMeasures()}")
print(f"  Notes: {strauss.getNumNotes()}")

# Visualize Chopin
print("\n📊 Chopin Fantaisie-Impromptu:")
ml.plotPianoRoll(chopin)

# Visualize Strauss
print("\n📊 Strauss Blue Danube Waltz:")
ml.plotPianoRoll(strauss)
````

**Comparative Analysis:**

**Chopin** (Romantic piano virtuoso piece):
- Very dense texture (many notes)
- Wide pitch range (uses full keyboard)
- Complex rhythmic patterns
- Continuous flow of notes

**Strauss** (Classical waltz for orchestra):
- More varied texture (some dense, some sparse sections)
- Clear separation between parts (melody, accompaniment)
- Regular rhythmic patterns (you can see the "oom-pah-pah" waltz pattern)
- More use of silence and space

This visual comparison immediately reveals stylistic differences between Romantic piano music and Classical orchestral dance music!

## Step 7: Understanding the Data Behind the Visualization

Piano rolls are created from the note data in the score. Let's see what data is being visualized.

````python
# Load a simple score
score = ml.Score('Bach_Cello_Suite_1.mxl')

# Convert to DataFrame to see the underlying data
df = score.toDataFrame()

print("Data used for piano roll visualization:")
print("\nFirst 20 notes:")
print(df[['measure', 'beat', 'pitch', 'midi', 'duration']].head(20))

# Calculate some statistics
print("\n📊 Score Statistics:")
print(f"  Pitch range: {df['midi'].min()} to {df['midi'].max()} ({df['pitch'].iloc[df['midi'].idxmin()]} to {df['pitch'].iloc[df['midi'].idxmax()]})")
print(f"  Average pitch: {df['midi'].mean():.1f} (≈ {ml.midiToPitch(int(df['midi'].mean()))})")
print(f"  Total duration: {df['duration'].sum():.1f} quarter notes")
````

**Understanding the Data:**

Each note in the piano roll is drawn using:

- **X position**: `measure + beat / beats_per_measure` (when the note starts)
- **Y position**: `midi` number (which pitch)
- **Width**: `duration` (how long the note lasts)
- **Height**: Fixed (just for visibility)
- **Color**: Based on part/instrument

This simple mapping of musical data to visual elements makes patterns immediately visible to the eye!

## Step 8: Analyzing Patterns in Piano Rolls

Let's use piano rolls to identify musical patterns visually.

````python
# Load Beethoven Symphony No. 5
beethoven = ml.Score('Beethoven_Symphony_5_mov_1.xml')

# Get just the first 20 measures to see the famous opening clearly
excerpt = ml.Score()
part = beethoven.getPart(0)  # First violins

excerpt_part = ml.Part()
excerpt_part.setPartName("First Violins (mm. 1-20)")

for i in range(min(20, part.getNumMeasures())):
    excerpt_part.addMeasure(part.getMeasure(i))

excerpt.addPart(excerpt_part)

# Visualize
ml.plotPianoRoll(excerpt)

print("\n🎵 Pattern Analysis:")
print("Look for:")
print("  1. The famous 'short-short-short-long' rhythmic motif")
print("  2. Repetitions of the motif at different pitch levels (sequences)")
print("  3. Rests (gaps in the visualization)")
print("  4. The gradual build-up of complexity")
````

**Musical Pattern Analysis:**

What you can see in this piano roll:

1. **Motivic repetition**: The same short pattern repeating
2. **Sequences**: The pattern moving to different pitch levels
3. **Development**: The pattern gets elaborated and extended
4. **Rhythmic consistency**: Even without hearing the music, you can see the rhythmic pattern visually

This is what makes Beethoven's Fifth Symphony so memorable - the obsessive repetition and development of a simple four-note motif!
