## Step 1: Import Maialib

Every Python script using Maialib starts the same way:

````python
# Import maialib and give it a shorter name 'ml'
# This is the standard way to import maialib
import maialib as ml

print(f"✅ Maialib {ml.__version__} loaded successfully!")
print("🎵 Ready to make music with code!")
````

**What just happened?**

- We imported the maialib library
- We used `as ml` to create a shorter alias (typing `ml.Note()` is easier than `maialib.Note()`)
- This is like opening your music software - now we're ready to create!

---

## Step 2: Create Your First Note

Let's create middle C (C4) - the most famous note in music!

````python
# Create a middle C note
note = ml.Note("C4")

# Print information about this note
print("Note Information:")
print(f"  Pitch: {note.getPitch()}")
print(f"  Octave: {note.getOctave()}")
print(f"  MIDI number: {note.getMIDI()}")
print(f"  Frequency: {note.getFrequency():.2f} Hz")
print(f"  Pitch class: {note.getPitchClass()}")
````

**Understanding the Output:**

- **Pitch**: The note name with octave (C4 = middle C)
- **Octave**: Which octave (4 = middle octave)
- **MIDI number**: Computer representation (60 = middle C, 61 = C#, 62 = D, etc.)
- **Frequency**: Sound wave frequency in Hertz (261.63 Hz for C4)
- **Pitch class**: Just the note name without octave (C, D, E, etc.)

---

## Step 3: Create More Notes

Let's create a few more notes and see how their properties differ:

````python
# Create notes for a C major scale
c = ml.Note("C4")
d = ml.Note("D4")
e = ml.Note("E4")
f = ml.Note("F4")
g = ml.Note("G4")
a = ml.Note("A4")
b = ml.Note("B4")
c_high = ml.Note("C5")

# Create a list to make it easier to work with
scale = [c, d, e, f, g, a, b, c_high]

# Print the scale
print("C Major Scale:")
print("Note  | MIDI | Frequency")
print("------|------|----------")
for note in scale:
    print(f"{note.getPitch():5} | {note.getMIDI():4} | {note.getFrequency():7.2f} Hz")
````

**Notice the Pattern:**

- MIDI numbers go up by 2 (whole step) or 1 (half step)
- Frequencies increase as notes get higher
- C5 is exactly one octave above C4 (double the frequency!)

---

## Step 4: Build Your First Chord

Now let's combine notes into a chord - three or more notes played together!

````python
# Create a C major chord (C-E-G)
c_major = ml.Chord(["C4", "E4", "G4"])

# Get information about the chord
print("C Major Chord:")
print(f"  Notes: {[note.getPitch() for note in c_major.getNotes()]}")
print(f"  Number of notes: {c_major.size()}")
print(f"  Chord name: {c_major.getName()}")
print(f"  Bass note (lowest): {c_major.getBass().getPitch()}")
print(f"  Is major? {c_major.isMajor()}")
print(f"  Is minor? {c_major.isMinor()}")
````

**Cool, right?** 🎉

Maialib automatically recognized that C-E-G is a C major chord! Let's try a minor chord:

````python
# Create an A minor chord (A-C-E)
a_minor = ml.Chord(["A3", "C4", "E4"])

print("A Minor Chord:")
print(f"  Notes: {[note.getPitch() for note in a_minor.getNotes()]}")
print(f"  Chord name: {a_minor.getName()}")
print(f"  Is major? {a_minor.isMajor()}")
print(f"  Is minor? {a_minor.isMinor()}")
````

## Step 5: Load a Real Musical Score

Now for the exciting part - let's load a real piece of classical music! Maialib comes with example scores, so you don't need to find your own files.

````python
# Load Bach's Cello Suite No. 1 (included with Maialib)
score = ml.Score('Bach_Cello_Suite_1.mxl')

print("✅ Score loaded successfully!")
print("\nScore Information:")
print(f"  Title: {score.getWorkTitle()}")
print(f"  Composer: {score.getComposer()}")
print(f"  Number of parts: {score.getNumParts()}")
print(f"  Number of measures: {score.getNumMeasures()}")
print(f"  Total notes: {score.getNumNotes()}")
````

**Amazing!** With just one line of code, you've loaded an entire musical score!

**Musical Context:** Bach's Cello Suite No. 1 is one of the most famous pieces ever written for solo cello. It has 543 notes across 42 measures. Now all of that musical information is in Python, ready to analyze!

---

## Step 6: Explore the Score

Let's look at the first few notes:

````python
# Get the first part (there's only one in this piece - the cello)
cello_part = score.getPart(0)

# Get the first measure
first_measure = cello_part.getMeasure(0)

# Get all notes from the first measure
notes = first_measure.getNotes()

# Print the first 10 notes
print("First 10 notes of Bach's Cello Suite No. 1:")
print("\nNote | Duration")
print("-----|----------")
for i, note in enumerate(notes[:10]):
    print(f"{note.getPitch():4} | {note.getDuration():.2f} quarter notes")
````

**Musical Note:** The piece starts with G, D, B, G - the opening of the famous Prelude. Each note is a sixteenth note (0.25 quarter notes).

---

## Step 7: Analyze Pitch Distribution

Let's do a simple analysis: which notes does Bach use most often?

````python
# Get all notes from the entire score
all_notes = score.getNotes()

# Count pitch classes (C, D, E, etc. - ignoring octave)
pitch_counts = {}
for note in all_notes:
    pitch_class = note.getPitchClass()
    if pitch_class in pitch_counts:
        pitch_counts[pitch_class] += 1
    else:
        pitch_counts[pitch_class] = 1

# Sort by count (most to least)
sorted_pitches = sorted(pitch_counts.items(), key=lambda x: x[1], reverse=True)

print("Note Usage in Bach Cello Suite No. 1:")
print("\nPitch | Count | Percentage")
print("------|-------|----------")
total = sum(pitch_counts.values())
for pitch, count in sorted_pitches:
    percentage = (count / total) * 100
    print(f"{pitch:5} | {count:5} | {percentage:5.1f}%")
````

**What This Tells Us:**

The most common notes are likely D, G, and B - the notes of the G major chord (the key of this suite). This makes sense because:

1. The suite is in G major
2. The cello's open strings are C, G, D, and A
3. Bach would naturally gravitate toward these notes

**You just did computational musicology!** 🎓

## Step 8: Your First Visualization

Let's create a simple visualization using Maialib's built-in plotting functions:

````python
# Create a piano roll visualization
# This shows notes as rectangles over time (like MIDI editors)
ml.plotPianoRoll(score)

print("🎨 Piano roll visualization created!")
print("(If running in Jupyter, you should see a graph above)")
````

**What you're seeing:**

- **X-axis**: Time (measures)
- **Y-axis**: Pitch (higher = higher notes)
- **Rectangles**: Individual notes (length = duration)

This gives you a bird's-eye view of the entire piece!
