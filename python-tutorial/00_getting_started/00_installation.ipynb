## Step 1: Check Python Installation

Maialib requires **Python 3.8 or newer**. Let's check if you have Python installed and which version.

````python
# Check Python version
import sys

print(f"Python version: {sys.version}")
print(f"Version info: {sys.version_info.major}.{sys.version_info.minor}.{sys.version_info.micro}")

# Check if version is compatible
if sys.version_info >= (3, 8):
    print("\n✅ Your Python version is compatible with Maialib!")
else:
    print("\n❌ Your Python version is too old. Please install Python 3.8 or newer.")
    print("Download from: https://www.python.org/downloads/")
````

**Note:** If you see an error running the cell above, Python might not be installed correctly. Visit [python.org](https://www.python.org/downloads/) to download and install Python.

## Step 3: Verify Installation

Let's make sure Maialib installed correctly by importing it and checking the version.

````python
# Import maialib
try:
    import maialib as ml
    print("✅ Maialib imported successfully!")
    print(f"📦 Version: {ml.__version__}")
    print(f"📍 Location: {ml.__file__}")
except ImportError as e:
    print("❌ Maialib import failed!")
    print(f"Error: {e}")
    print("\nTry reinstalling with: pip install maialib --upgrade")
````

If you see "✅ Maialib imported successfully!", congratulations! You're ready to start making music with code! 🎉

---

## Step 4: Quick Test

Let's create your first musical note to make sure everything works!

````python
# Create a middle C note
note = ml.Note("C4")

# Print information about the note
print(f"Note: {note.getPitch()}")
print(f"MIDI Number: {note.getMIDI()}")
print(f"Frequency: {note.getFrequency():.2f} Hz")
print(f"Pitch Class: {note.getPitchClass()}")

print("\n🎵 If you see this output, Maialib is working perfectly!")
````

## Troubleshooting Common Issues

### Issue 1: "pip: command not found"

**Solution:** Python might not be in your system PATH. Try:

```bash
python -m pip install maialib
# or
python3 -m pip install maialib
```

### Issue 2: "ModuleNotFoundError: No module named 'maialib'"

**Solutions:**

1. Make sure you're using the same Python environment where you installed maialib
2. Try reinstalling:
   ```bash
   pip uninstall maialib
   pip install maialib
   ```
3. Check if maialib is installed:
   ```bash
   pip list | grep maialib
   ```

### Issue 3: Permission Errors

**Solution:** You might need administrator/sudo permissions:

```bash
# On Mac/Linux
sudo pip install maialib

# Or install just for your user (no admin needed)
pip install --user maialib
```

### Issue 4: "No matching distribution found"

**Solution:** Your Python version might be too old or too new. Check compatibility:

- Maialib supports Python 3.8 through 3.14
- Upgrade or downgrade Python if needed

## 🎯 Practice Exercise

Try creating a few different notes and exploring their properties:

````python
# Create three notes: C4, E4, G4 (a C major chord)
c = ml.Note("C4")
e = ml.Note("E4")
g = ml.Note("G4")

# Print their frequencies
print("C major triad frequencies:")
print(f"C4: {c.getFrequency():.2f} Hz")
print(f"E4: {e.getFrequency():.2f} Hz")
print(f"G4: {g.getFrequency():.2f} Hz")

# Calculate the frequency ratios (important in music theory!)
print(f"\nFrequency ratios:")
print(f"E4/C4 = {e.getFrequency()/c.getFrequency():.4f} (major third ≈ 1.25)")
print(f"G4/C4 = {g.getFrequency()/c.getFrequency():.4f} (perfect fifth ≈ 1.5)")
````

**Musical Note:** Notice how the frequency ratios approximate simple fractions (5/4 and 3/2)! This is the foundation of consonance in Western music theory. In equal temperament tuning, these ratios are slightly adjusted, but they're very close to these "pure" intervals.
